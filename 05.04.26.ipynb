Support Vector Machine (SVM)

Support Vector Machine (SVM) is a machine learning algorithm that separates data into classes by drawing the best possible boundary (line or plane) between them.

📘 বাংলায় (সহজভাবে):
SVM হলো এমন একটি অ্যালগরিদম যা ডেটাকে আলাদা করার জন্য সবচেয়ে ভালো লাইন বা বাউন্ডারি তৈরি করে।


Support Vector Machine (SVM) is a supervised machine learning algorithm used mainly for:

Classification (e.g., spam vs not spam)
Regression (predicting values)

👉 It finds the best boundary (hyperplane) that separates data into different classes.

📘 বাংলায়:
SVM হলো একটি মেশিন লার্নিং অ্যালগরিদম যা ডেটাকে আলাদা আলাদা ক্লাসে ভাগ করার জন্য সবচেয়ে ভালো সীমারেখা (boundary) খুঁজে বের করে।

SVM tries to maximize the margin between two classes
The closest points to the boundary are called Support Vectors

📘 বাংলায়:
SVM এমন একটি লাইন বা প্লেন খুঁজে যা দুই ক্লাসের মধ্যে সবচেয়ে বেশি দূরত্ব (margin) তৈরি করে।
যে পয়েন্টগুলো এই লাইনের সবচেয়ে কাছে থাকে, সেগুলোই support vectors।

How it Works (কিভাবে কাজ করে?)
Find a separating line (in 2D) or plane (in higher dimensions)
Maximize the margin between classes
Use only critical points (support vectors)

📘 বাংলায়:

ডেটাকে আলাদা করার জন্য একটি লাইন/প্লেন খুঁজে
দুই ক্লাসের মধ্যে দূরত্ব বাড়ায়
শুধু গুরুত্বপূর্ণ পয়েন্ট (support vectors) ব্যবহার করে

Geting Started with SVM Classification

```python
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import LabelEncoder
```

Digits Loading

```python
digits = datasets.load_digits()
#print(digits)
images_and_labels = list(zip(digits.images,digits.target))
```

```python
digits.images
```

```python
digits.target
```

Flattening the Images

```python
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))
```

Training the Model

```python

#Fetching training set, 75% of entire dataset, rest will be used for testing

trainTestSplit = int(n_samples*0.75)

#Training the model
model = svm.SVC()
model.fit(data[:trainTestSplit], digits.target[:trainTestSplit])

target = digits.target[trainTestSplit:]
predictions = model.predict(data[trainTestSplit:])
```

Printing an Error Matrix and Accuracy Score

```python

print(confusion_matrix(target,predictions))
print(accuracy_score(target,predictions))
```

```python

import seaborn as sns
sns.heatmap(confusion_matrix(target,predictions), annot=True, cmap='Blues')
```

Prediction

```python
plt.imshow(digits.images[-55], cmap = plt.cm.gray_r, interpolation = "nearest")
print("Prediction for test image:", model.predict(data[-55].reshape(1,-1)))

plt.show()
```

**Decision Tree Classification (Easy Way)

Decision Tree is a machine learning algorithm that makes decisions like a tree (flowchart style).

👉 It asks questions step-by-step to classify data.

📘 বাংলায় (সহজভাবে):
Decision Tree হলো এমন একটি অ্যালগরিদম যা ধাপে ধাপে প্রশ্ন করে সিদ্ধান্ত নেয় (গাছের মতো structure)।

👉 Think like this:
Is age < 18?
   ├── Yes → Student
   └── No → Working

✔ Each question = Node
✔ Each branch = Decision path
✔ Final result = Leaf node

📘 বাংলায়:
প্রতিটা প্রশ্ন = node
প্রতিটা রাস্তা = branch
শেষ ফলাফল = leaf

⚙️ How it Works (কিভাবে কাজ করে?)
Pick best feature (question)
Split data into groups
Repeat until clear decision

📘 বাংলায়:

সবচেয়ে ভালো প্রশ্ন নির্বাচন করে
ডেটা ভাগ করে
বারবার করে যতক্ষণ না clear result আসে
🔍 Example (Real Life)

👉 Suppose:

Weather = Sunny / Rainy
Play? Yes / No

Tree:

Is weather sunny?
   ├── Yes → Play
   └── No → Don’t play

📘 বাংলায়:
আবহাওয়া দেখে সিদ্ধান্ত নেয় খেলবে কি না
