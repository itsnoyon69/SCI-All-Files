Method in List

```python
c=[6,7,8,9,10]

c.append(11)        #append a item
print(c)

c.insert(2,12)      #insert 12 in 2no index
print(c)

c.pop()             #pop/remove last item
print(c)

c.pop(2)            #remove ith item
print(c)

del c[2]            #delete ith item
print(c)

print(c.index(6))    #show item index no

print(c.count(6))   #count the item has
```

Method in string

```python
d="Noyon"
d.center(10)            #center in d
print(d)

print(d.count("o"))     #count item

d.ljust(15)
print(d)

print(d.split("o"))  #split all with item

print(d.find("o"))      #find the item index no and r/l is size where it will start

```

Set Operation

```python
a={1,2,3}
b={1,3,5,7}
print(2 in a)   #Check item has ot Not

print(a | b)      #Union Set

print(a & b)        #only common

print(a<=b)         #Subset Check
```

Method In Set

```python
x={5,6,7,8}
y={5,7,9,0}

print(x.union(y))          #union x,y
print(x.intersection(y))   #intersection
print(x.difference(y))      #x-y
print(x.issubset(y))        #subset check
x.add(10)                   #add 10 in set
print(x)
x.remove(10)                #Remove 10 from set
print(x)
x.clear()                   #clear the set
print(x)
```

Dictionary

```python
myDict={"5201":"Noyon","6058":"Shifat"}
print(myDict)           #Print dict
myDict["4679"]="Shaon"
print(myDict)               #Add in Dict
for i in myDict:
    print(myDict[i],"'s ID is",i)
del myDict["4679"]          #Delete from Key
print(myDict)
```

Method in dictionary

```python
print(myDict.keys())        #Show all keys of Dict
print(myDict.values())      #Show all values of Dict
print(myDict.items())       #Show all in pair
print(myDict.get("5201"))   #Show the value of this key
print(myDict.get("4679","Shaon")) #Show the value
```
