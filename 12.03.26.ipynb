HackerRank String 1

```python
def minion_game(string):
    s = string.upper()
    count1 = 0
    count2 = 0
    l = len(s)

    for i in range(l):
        if s[i] in "AEIOU":
            count1 = count1 + l - i
        else:
            count2 = count2 + l - i

    if count1 > count2:
        print("Kevin", count1)
    elif count2 > count1:
        print("Stuart", count2)
    else:
        print("Draw")


if __name__ == '__main__':
    s = str(input())
    minion_game(s)
```

HackerRank String 2

```python
def merge_the_tools(string, k):

    n = len(string)

    for i in range(0, n, k):
        temp = string[i:i+k]
        temp1 = ""

        for j in temp:
            if j not in temp1:
                temp1 = temp1 + j

        print(temp1)


if __name__ == '__main__':
    string, k = input(), int(input())
    merge_the_tools(string, k)
```

HackerRank Error Handling

```python
t = int(input())

for _ in range(t):
    try:
        a, b = input().split()
        print(int(a) // int(b))
    except ZeroDivisionError:
        print("Error Code: integer division or modulo by zero")
    except ValueError as e:
        print("Error Code:", e)
```

HackerRank Regex

```python
import re

t = int(input())

for _ in range(t):
    a = input()
    try:
        re.compile(a)
        print(True)
    except re.error:
        print(False)
```
