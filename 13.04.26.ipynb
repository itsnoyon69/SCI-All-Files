What is Neural Network?

English:
Neural Network is a machine learning model inspired by the human brain that learns patterns from data.

Bangla:
Neural Network аҰ№аҰІа§Ӣ аҰ®аҰҫаҰЁа§ҒаҰ·а§ҮаҰ° аҰ®аҰёа§ҚаҰӨаҰҝаҰ·а§ҚаҰ•а§ҮаҰ° аҰ®аҰӨа§Ӣ аҰҸаҰ•аҰҹаҰҝ model, аҰҜаҰҫ data аҰҰа§ҮаҰ–а§Ү pattern аҰ¶аҰҝаҰ–а§Ү

рҹ”№ Simple Idea

рҹ‘ү Input вҶ’ Process вҶ’ Output

рҹ‘ү Example:

image вҶ’ network вҶ’ вҖңcatвҖқ

data вҶ’ network вҶ’ prediction

рҹ”№ How it works (super easy)

Input вҶ’ Weight вҶ’ Sum вҶ’ Activation вҶ’ Output

рҹ‘ү Bangla:

input аҰЁа§Үа§ҹ

weight аҰҰаҰҝа§ҹа§Ү аҰ№аҰҝаҰёаҰҫаҰ¬ аҰ•аҰ°а§Ү

decision аҰҰа§Үа§ҹ

рҹ”‘ Key Points

weight = importance

activation = decision function

learning = error аҰ•аҰ®аҰҫаҰЁа§Ӣ

Activation Function
вң… What (аҰ•аҰҝ?)

English:
Activation Function decides neuron output

Bangla:
Activation Function аҰ аҰҝаҰ• аҰ•аҰ°а§Ү neuron аҰ•аҰҫаҰң аҰ•аҰ°аҰ¬а§Ү аҰ•аҰҝаҰЁаҰҫ

рҹҺҜ Main Work (аҰ•аҰҫаҰң)

output decide аҰ•аҰ°а§Ү

non-linearity аҰҰа§Үа§ҹ

learning possible аҰ•аҰ°а§Ү

рҹ“Ҡ Types of Activation Function

вӯҗ 1. ReLU

рҹ‘ү Definition

f(x) = max(0, x)

рҹ‘ү аҰ•аҰҫаҰң
negative вҶ’ 0
positive вҶ’ same

рҹ‘ү Use
hidden layer
most popular

рҹ”№ 2. Sigmoid

рҹ‘ү Definition
f(x) = 1 / (1 + e^-x)

рҹ‘ү аҰ•аҰҫаҰң
output: 0 to 1

рҹ‘ү Use
binary classification

рҹ”№ 3. Tanh

рҹ‘ү Definition
f(x) = tanh(x)

рҹ‘ү аҰ•аҰҫаҰң
output: -1 to 1

рҹ‘ү Use
hidden layer

рҹ”№ 4. Softmax

рҹ‘ү Definition
рҹ‘ү probability function

рҹ‘ү аҰ•аҰҫаҰң
multiple class probability аҰҰа§Үа§ҹ
рҹ‘ү Use
multi-class classification

рҹ”№ 5. Linear
рҹ‘ү Definition
f(x) = x
рҹ‘ү аҰ•аҰҫаҰң
straight output
рҹ‘ү Use
regression

рҹ”№ 6. Step Function
рҹ‘ү Definition
if x вүҘ threshold вҶ’ 1 else 0
рҹ‘ү аҰ•аҰҫаҰң
yes / no decision
рҹ‘ү Use
simple binary

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras

# 1. Model create
model = tf.keras.Sequential([
    keras.layers.Dense(units=1, input_shape=[1])
])

# 2. Compile model
model.compile(optimizer='sgd', loss='mean_squared_error')

# 3. Data аҰҰа§ҮаҰ“а§ҹаҰҫ
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)

# 4. Train model
model.fit(xs, ys, epochs=500)

# 5. Predict
print(model.predict(np.array([10.0])))
```

Gradient Descent Algorithm

What is Gradient Descent?

English:
Gradient Descent is an algorithm used to minimize loss (error)

Bangla:
Gradient Descent аҰ№аҰІа§Ӣ error аҰ•аҰ®аҰҫаҰЁа§ӢаҰ° аҰҸаҰ•аҰҹаҰҝ algorithm

рҹ”№ Simple Idea

рҹ‘ү аҰӘаҰҫаҰ№аҰҫа§ң аҰҘа§ҮаҰ•а§Ү аҰЁаҰҝаҰҡа§Ү аҰЁаҰҫаҰ®аҰҫаҰ° аҰ®аҰӨа§Ӣ
рҹ‘ү always go downhill (minimum)

вҡҷпёҸ Main Formula
theta = theta - alpha * grad

рҹ‘ү Bangla:

theta = weight

alpha = learning rate (step size)

grad = slope

рҹ”Ғ How it works (Easy)

random value аҰҰаҰҝа§ҹа§Ү аҰ¶а§ҒаҰ°а§Ғ
gradient аҰ¬а§ҮаҰ° аҰ•аҰ°а§Ү

update аҰ•аҰ°а§Ү

repeat

рҹ”‘ Important Rules

negative gradient follow аҰ•аҰ°аҰӨа§Ү аҰ№аҰ¬а§Ү

learning rate аҰ–а§ҒаҰ¬ аҰ¬а§ң аҰ№аҰІа§Ү аҰӯа§ҒаҰІ аҰ№аҰ¬а§Ү

learning rate аҰӣа§ӢаҰҹ аҰ№аҰІа§Ү slow аҰ№аҰ¬а§Ү


Gradient Descent Algorithm (Step by Step)
рҹ§  Algorithm (Easy)

Step 1: Initialize

рҹ‘ү random value аҰҰаҰҝа§ҹа§Ү аҰ¶а§ҒаҰ°а§Ғ (theta)

Step 2: Calculate Prediction

рҹ‘ү model output аҰ¬а§ҮаҰ° аҰ•аҰ°а§Ү

Step 3: Calculate Error (Loss)

рҹ‘ү actual vs predicted compare

Step 4: Calculate Gradient

рҹ‘ү slope аҰ¬а§ҮаҰ° аҰ•аҰ°аҰҫ (error аҰ•аҰӨ аҰҰаҰҝаҰ•а§Ү аҰ•аҰ®аҰ¬а§Ү)

Step 5: Update Weight
theta = theta - alpha * gradient

Step 6: Repeat

рҹ‘ү аҰ…аҰЁа§ҮаҰ•аҰ¬аҰҫаҰ° repeat (epochs)

рҹҺҜ Algorithm Short Form
Start вҶ’ Predict вҶ’ Error вҶ’ Gradient вҶ’ Update вҶ’ Repeat
рҹ“Ң Now Example (Step by Step)
рҹҺҜ Problem

аҰ§аҰ°а§Ӣ relation:

Y = 3X + 1

аҰҶаҰ®аҰ°аҰҫ model train аҰ•аҰ°аҰ¬а§Ӣ

рҹ”№ Step 1: Initialize

рҹ‘ү аҰ§аҰ°а§Ӣ model guess аҰ•аҰ°а§Ү:

Y = 5X + 5 вқҢ
рҹ”№ Step 2: Data
X = 1 вҶ’ Y = 4
рҹ”№ Step 3: Prediction

рҹ‘ү model аҰ¬аҰІаҰІа§Ӣ:

Y = 5(1) + 5 = 10
рҹ”№ Step 4: Error
Error = 10 - 4 = 6

рҹ‘ү Bangla:
рҹ‘ү model аҰ…аҰЁа§ҮаҰ• аҰӯа§ҒаҰІ аҰ•аҰ°аҰӣа§Ү

рҹ”№ Step 5: Gradient

рҹ‘ү gradient аҰ¬аҰІа§Ү:
рҹ‘ү weight аҰ•аҰ®аҰҫаҰӨа§Ү аҰ№аҰ¬а§Ү

рҹ”№ Step 6: Update
theta = theta - alpha * gradient

рҹ‘ү аҰ§аҰ°а§Ӣ аҰЁаҰӨа§ҒаҰЁ model:

Y = 4X + 3 (better)
рҹ”№ Step 7: Repeat

рҹ‘ү аҰҶаҰ¬аҰҫаҰ° error аҰ•аҰ®аҰ¬а§Ү

рҹ‘ү аҰ¶а§ҮаҰ·а§Ү:

Y вүҲ 3X + 1 вң…

What is Cost Function?

English:
Cost Function measures how wrong the model is

Bangla:
Cost Function model аҰ•аҰӨаҰҹаҰҫ аҰӯа§ҒаҰІ аҰ•аҰ°аҰӣа§Ү аҰӨаҰҫ аҰ®аҰҫаҰӘа§Ү

рҹ”№ Simple Idea

рҹ‘ү predicted vs actual compare аҰ•аҰ°а§Ү

рҹ‘ү difference = error

рҹ‘ү аҰёаҰ¬ error аҰҸаҰ° average = cost

вҡҷпёҸ Formula Idea

Cost = average of all errors

рҹ”‘ Key Point
Loss = single data error
Cost = all data average error
рҹ“Ң Example

аҰ§аҰ°а§Ӣ:

X	Actual Y	Predicted Y
1	4	6
2	7	8
3	10	9

рҹ‘ү Error:

(6-4)=2
(8-7)=1
(9-10)=1

рҹ‘ү Cost:

(2+1+1)/3 = 1.33
рҹҺҜ Why Important?

рҹ‘ү Gradient Descent аҰҸаҰҮ cost аҰ•аҰ®аҰҫа§ҹ

рҹ‘ү goal:

Cost вҶ“ вҶ’ Model better

рҹ§  Real Life Idea

рҹ‘ү exam marks

аҰӨа§ӢаҰ®аҰҫаҰ° аҰӯа§ҒаҰІ аҰ¬а§ҮаҰ¶аҰҝ вҶ’ score аҰ–аҰҫаҰ°аҰҫаҰӘ
аҰӯа§ҒаҰІ аҰ•аҰ® вҶ’ score аҰӯаҰҫаҰІа§Ӣ

рҹ‘ү same as cost function

SGD Algorithm (Stochastic Gradient Descent)

English:
SGD is an algorithm that updates model weights using one random data point at a time to reduce error
