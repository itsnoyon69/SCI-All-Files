Regex All

```
import re

text = "I have 2 apples and 15 oranges"

numbers = re.findall(r'\d+', text) #Find all numbers

print(numbers)
```

```
import re

text = "Emails: test806@gmail.com, user@yahoo23.com"

emails = re.findall(r'\w+\d+@\w+\.\w+', text)   #Email find korbe

print(emails)
```

```
import re

text = "I like cat"

new_text = re.sub(r'cat', 'dog', text)      #Replace korbe

print(new_text)
```

```
import re

password=input()

pattern = r'[a-z][A-Z]+\d+'      #Validate Check

if re.match(pattern, password):
    print("Valid password")
else:
    print("Invalid")
```

```
import re

text = "Today is 2026-03-15"

date = re.findall(r'\d{4}-\d{2}-\d{2}', text)         #Date diye exact number khuja

print(date)
```

Error Exception

```
try:
    num = int(input("Enter a number: "))
    print("Valid number:",num)      #Specific Check

except ValueError:
    print("Invalid number")
```

```
try:
    a = int(input())
    b = int(input())
    print(a / b)

except ZeroDivisionError:
    print("Cannot divide by zero")      #Multiple error

except ValueError:
    print("Invalid input")
```

```
try:
    a = 10
    b = 0
    print(a / b)

except ZeroDivisionError:
    print("0 দিয়ে ভাগ করা যায় না")

else:
    print("Division successful")        #except kaj na korle

finally:
    print("Program finished")
```

```
try:
    num = int(input("Enter number: "))      #ValueError
    print(num)

except ValueError:
    print("Number ছাড়া অন্য কিছু দেওয়া হয়েছে")

else:
    print("Input correct")

finally:
    print("Done")
```

```
try:
    result = 5 + "10"
    print(result)                   #TypeError

except TypeError:
    print("Different data type add করা যাবে না")

finally:
    print("Finished")
```

Class Inheritance

```
class Animal:
    def sound(self):
        print("Animal sound")

class Dog(Animal):
    pass

d = Dog()
d.sound()
```

Polymorphism

```
class Bird:
    def fly(self):
        print("Bird can fly")

class Penguin:
    def fly(self):
        print("Penguin cannot fly")

b = Bird()
p = Penguin()

b.fly()
p.fly()
```

Encapsulation

```
class Student:
    def __init__(self):
        self.__marks = 80

    def show(self):
        print(self.__marks)

s = Student()
s.show()
```
