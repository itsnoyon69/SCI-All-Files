Class Inheritance

```python
class Vehicle:
    def __init__(self,name):
        self.name = name
    def show(self):
        print(self.name)
class Car(Vehicle):
    def speed(self):
        print(self.name,"has better speed")
class Bike(Vehicle):
    def sound(self):
        print(self.name,"has better sound")
name=input()
c=Car(name=input())
c.speed()
c.show()
b=Bike(name=input())
b.sound()
b.show()
```

Polymorphism

```python
class Dog:
    def sound(self,name):
        self.name=name
        print(self.name,"says hi")
class Cat:
    def sound(self,name):
        self.name=name
        print(self.name,"says hellow")

d=Dog()
c=Cat()
d.sound("Tom")
c.sound("PewMew")
```

Encapsulation

```python
class User:
    def __init__(self,pas):
        self.__pas=pas
    def check_pass(self,p):
        if p==self.__pas:
            print("Logged in")
        else:
            print("Wrong password")

u=User("Noyon")
p=input("enter your password")
u.check_pass(p)
```

Abstract Class

```python
from abc import ABC, abstractmethod

class ATM(ABC):         #ABC Class

    @abstractmethod
    def withdraw(self):     #Virtual Method
        pass


class BankATM(ATM):
    def withdraw(self):
        print("Money withdrawn")


b = BankATM()
b.withdraw()
```

Pure Virtual Function (Abstract Method)

```python
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass
```

Interface
class InterfaceName(ABC)
       ↓
@abstractmethod
       ↓
Child class must implement it

```python
from abc import ABC, abstractmethod

# Interface
class Vehicle(ABC):         #interface

    @abstractmethod
    def start(self):
        pass


# Implementation class
class Car(Vehicle):

    def start(self):
        print("Car started")


c = Car()
c.start()
```

XML

```python
import xml.etree.ElementTree as etree
n=int(input())
xml=""
for i in range(n):
    xml+=input()
tree=etree.ElementTree(etree.fromstring(xml))
root=tree.getroot()
count=0
for elem in root.iter():
    count+=len(elem.attrib)
print(count)
```

Debugging

```python
def score_words():
    n = int(input())
    words = input().split()

    total = 0

    for i in words:
        count = 0

        for j in i:
            if j in "aeiouy":
                count += 1

        if count % 2 == 0:
            total += 2
        else:
            total += 1

    return total


print(score_words())
```
