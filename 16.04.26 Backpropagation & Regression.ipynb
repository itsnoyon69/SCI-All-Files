👉 Definition

simple binary

Backpropagation

English:
Backpropagation is an algorithm used to train neural networks by adjusting weights based on error.

বাংলা:
Backpropagation হলো একটি অ্যালগরিদম যা নিউরাল নেটওয়ার্কে ভুল (error) অনুযায়ী weight ঠিক করে শেখায়।

🎯 Why it is used?

English:

To reduce error
To improve prediction accuracy
To train deep learning models

বাংলা:

ভুল কমানোর জন্য
সঠিক ফল পাওয়ার জন্য
মডেল ট্রেন করার জন্য

⚙️ How it works (Step-by-step)
1. Forward Pass

English: Input goes through network → output comes
বাংলা: ইনপুট দিয়ে আউটপুট বের করা হয়

2. Loss Calculation

English: Compare output with actual value
বাংলা: আসল মানের সাথে তুলনা করে ভুল বের করা হয়

3. Backward Pass

Uses Chain Rule

English: Error goes backward and gradients are calculated
বাংলা: ভুল পিছনের দিকে পাঠিয়ে gradient বের করা হয়

4. Weight Update

English: Update weights using gradient descent
বাংলা: weight আপডেট করা হয়

🧮 Important Formula

∂L/∂w

English: Change in loss with respect to weight
বাংলা: weight পরিবর্তনে loss কতটা বদলায়

💻 Syntax (Conceptual)
for data in dataset:
    output = model(input)
    loss = loss_function(output, target)
    loss.backward()   # backpropagation
    optimizer.step()  # update weights



Simple Code

```python
import numpy as np

# Activation function (Sigmoid)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of sigmoid
def sigmoid_derivative(x):
    return x * (1 - x)

# Input (X) and Output (y)
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

y = np.array([[0],
              [1],
              [1],
              [0]])

# Initialize weights randomly
np.random.seed(42)
weights_input_hidden = np.random.rand(2, 2)
weights_hidden_output = np.random.rand(2, 1)

# Training
lr = 0.1
epochs = 10000

for epoch in range(epochs):

    # 🔹 Forward Pass
    hidden_layer_input = np.dot(X, weights_input_hidden)
    hidden_layer_output = sigmoid(hidden_layer_input)

    final_input = np.dot(hidden_layer_output, weights_hidden_output)
    final_output = sigmoid(final_input)

    # 🔹 Error
    error = y - final_output

    # 🔹 Backpropagation
    d_output = error * sigmoid_derivative(final_output)

    error_hidden = d_output.dot(weights_hidden_output.T)
    d_hidden = error_hidden * sigmoid_derivative(hidden_layer_output)

    # 🔹 Update Weights (SGD style)
    weights_hidden_output += hidden_layer_output.T.dot(d_output) * lr
    weights_input_hidden += X.T.dot(d_hidden) * lr

# Output after training
print("Final Output:\n", final_output)
```

```python
horsepower = np.array(train_features['Horsepower'])

horsepower_normalizer = Normalization(input_shape=[1,])
horsepower_normalizer.adapt(horsepower)
```

```python
horsepower = np.array(train_features['Horsepower']).reshape(-1,1)

horsepower_normalizer = Normalization()
horsepower_normalizer.adapt(horsepower)
```

```python
horsepower_model = tf.keras.Sequential([
    horsepower_normalizer,
    layers.Dense(1)
])
```

```python
horsepower_model.predict(horsepower[:10])
```

```python
horsepower_model.compile(
    optimizer=tf.optimizers.Adam(learning_rate=0.1),
    loss='mean_absolute_error')
```

```python
history = horsepower_model.fit(
    train_features['Horsepower'], train_labels,
    epochs=100,
    # suppress logging
    verbose=0,
    # Calculate validation results on 20% of the training data
    validation_split = 0.2)
```

Visualize the model's training progress using the stats stored in the history object.

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
def plot_loss(history):
  plt.plot(history.history['loss'], label='loss')
  plt.plot(history.history['val_loss'], label='val_loss')
  plt.ylim([0, 10])
  plt.xlabel('Epoch')
  plt.ylabel('Error [MPG]')
  plt.legend()
  plt.grid(True)
```

```python
plot_loss(history)
```

```python
test_results = {}

test_results['horsepower_model'] = horsepower_model.evaluate(
    test_features['Horsepower'],
    test_labels, verbose=0)
```

A DNN regression

```python
def build_and_compile_model(norm):
  model = keras.Sequential([
      norm,
      layers.Dense(64, activation='relu'),
      layers.Dense(64, activation='relu'),
      layers.Dense(1)
  ])

  model.compile(loss='mean_absolute_error',
                optimizer=tf.keras.optimizers.Adam(0.001))
  return model
```

Conclusion


This notebook introduced a few techniques to handle a regression problem. Here are a few more tips that may help:

Mean Squared Error (MSE) and Mean Absolute Error (MAE) are common loss functions used for regression problems. Mean Absolute Error is less sensitive to outliers. Different loss functions are used for classification problems.
Similarly, evaluation metrics used for regression differ from classification.
When numeric input data features have values with different ranges, each feature should be scaled independently to the same range.
