Introduce to ML

What is Machine Learning?

Machine Learning is a technology that allows computers to learn from data and improve automatically without being explicitly programmed.

Machine Learning is a subset of Artificial Intelligence (AI).
ML = Learn from data + Improve automatically

Example:If you give many images of cats, the computer learns to recognize cats.

Machine learning Process

1- Define the objective of the Problem
2- Data Gathering
3- Data Preprocessing
4- Exploratory Data Analysis
5- Building a Machine Learning Model
6- Model Evaluation & Optimization
7- Predictions
8- Building a Model Package for reusability

Types of Machine Learning

1.Supervised Learning is a
technique in which we teach or
train the machine using labelled
data

2.Unsupervised learning is the
training of machine using
information that is unlabeled and
allowing the algorithm  to act on
that information without
guidance

3.Reinforcement Learning is a part of
Machine Learning where an agent is
put in an environment and he learns
to behave in this environment by
performing certain actions and
observing  the rewards which it gets
from those actions

StandardScaler: StandardScaler is used to scale data so that it has:

Mean = 0
Standard Deviation = 1

Different features may have different scales (e.g., age vs salary). Scaling makes them equal.

from sklearn.preprocessing import StandardScaler

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

data = np.array([[10], [20], [30], [40]])

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

print(scaled_data)
```

MinMaxScaler: MinMaxScaler scales data into a fixed range, usually 0 to 1

It makes all values small and within the same range.

📌 Example:
10 → 0.0
20 → 0.33
30 → 0.66
40 → 1.0

🔹 Import
from sklearn.preprocessing import MinMaxScaler

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

data = np.array([[10], [20], [30], [40]])

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)

print(scaled_data)
```

RobustScaler: RobustScaler scales data using median and interquartile range (IQR),so it is not affected by outliers.

If your data has outliers, RobustScaler works better than other scalers.

📌 Example:
Data: [10, 20, 30, 1000] → এখানে 1000 একটি outlier

🔹 Import
from sklearn.preprocessing import RobustScaler

```python
import numpy as np
from sklearn.preprocessing import RobustScaler

data = np.array([[10], [20], [30], [1000]])

scaler = RobustScaler()
scaled_data = scaler.fit_transform(data)

print(scaled_data)
```

Normalizer: Normalizer scales each data row (sample) so that its length becomes 1 (unit norm)

It works row-wise (each sample), not column-wise

🔹 Why use it? (কেন ব্যবহার করি)

Useful when direction matters more than magnitude (e.g., text data)

যখন মানের আকার নয়, বরং দিক (pattern) গুরুত্বপূর্ণ হয় (যেমন: text data)

🔹 Import
from sklearn.preprocessing import Normalizer

```python
import numpy as np
from sklearn.preprocessing import Normalizer

data = np.array([[3, 4, 5]])

scaler = Normalizer()
normalized_data = scaler.fit_transform(data)

print(normalized_data)
```

Custom Transformer : A Custom Transformer is a user-defined tool to transform data in your own way.


🔹 Why use it? (কেন ব্যবহার করি)

When built-in tools (Scaler, Encoder) are not enough

👉 বাংলা:
যখন ready-made tools (Scaler, Encoder) দিয়ে কাজ হয় না

🔹 Key Idea (মূল ধারণা)

👉 English:
You create your own transformation logic

🔹 Import Needed
from sklearn.base import BaseEstimator, TransformerMixin

```python
from sklearn.base import BaseEstimator, TransformerMixin
import numpy as np

class CustomTransformer(BaseEstimator, TransformerMixin):

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        return X * 2   # example: multiply data by 2

data = np.array([[1], [2], [3]])

ct = CustomTransformer()
result = ct.fit_transform(data)

print(result)
```

Text → Count words → Numbers

from sklearn.feature_extraction.text import CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

texts = ["I love ML", "ML is fun"]

cv = CountVectorizer()
X = cv.fit_transform(texts)

print(X.toarray())
```

TF-IDF Vectorizer : TF-IDF Vectorizer converts text into numbers based on importance of words (not just count)

👉 বাংলা:
TF-IDF Vectorizer লেখা কে সংখ্যায় রূপান্তর করে, কিন্তু শুধু গোনা না—শব্দের গুরুত্ব অনুযায়ী

🔹 Full Meaning

TF = Term Frequency (how often word appears)
IDF = Inverse Document Frequency (how rare the word is)

TF = শব্দ কতবার এসেছে
IDF = শব্দ কতটা rare (কমন না)

🔹 Simple Idea

👉 English:
Important words → higher value
Common words → lower value

👉 বাংলা:
গুরুত্বপূর্ণ শব্দ → বেশি মান
সাধারণ শব্দ → কম মান

🔹 Import
from sklearn.feature_extraction.text import TfidfVectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer

texts = ["I love ML", "ML is fun"]

tfidf = TfidfVectorizer()
X = tfidf.fit_transform(texts)

print(X.toarray())
```

HashingVectorizer: HashingVectorizer converts text into numbers using a hash function (no need to store vocabulary)

HashingVectorizer লেখা কে সংখ্যায় রূপান্তর করে একটি hash function ব্যবহার করে (শব্দের তালিকা আলাদা করে রাখতে হয় না)

🔹 Simple Idea (সহজ ধারণা)

👉 English:
Text → Hash → Fixed-size numbers

👉 বাংলা:
Text → Hash → নির্দিষ্ট আকারের সংখ্যা

🔹 Import
from sklearn.feature_extraction.text import HashingVectorizer

```python
from sklearn.feature_extraction.text import HashingVectorizer

texts = ["I love ML", "ML is fun"]

hv = HashingVectorizer(n_features=5)
X = hv.transform(texts)

print(X.toarray())
```
