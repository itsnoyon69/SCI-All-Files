Bar Plot: Compares different categories

Pie Chart: English:

Stack Plot :
A Stack Plot shows how multiple data values combine over time (stacked on top of each other)

👉 বাংলা:
Stack Plot দেখায় কিভাবে একাধিক ডেটা একসাথে যোগ হয়ে সময় অনুযায়ী পরিবর্তন হয়

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]

y1 = [2, 3, 4, 5]
y2 = [1, 2, 1, 2]
y3 = [3, 1, 2, 3]

plt.stackplot(x, y1, y2, y3, labels=['A', 'B', 'C'])

plt.title("Stack Plot")
plt.xlabel("X Axis")
plt.ylabel("Values")
plt.legend()

plt.show()
```

Simple Linear Regression

Regression & Correlation
🔹 What is Regression?

Regression is a method to find the relationship between one output (Y) and one or more inputs (X)

📌 Example:
Study hours → Marks

🔹 What is Correlation?

👉 English:
Correlation measures how strongly two variables are related

👉 বাংলা:
Correlation বলে দুইটি ভেরিয়েবল কতটা একসাথে পরিবর্তন হয়


🔹 Important Statement

👉 English:
If there is no correlation, then regression cannot find a meaningful relationship

👉 বাংলা:
যদি correlation না থাকে, তাহলে regression কোনো meaningful সম্পর্ক বের করতে পারে না

🔹 Why? (কেন?)

👉 English:
Regression needs a pattern (relationship) to learn

👉 বাংলা:
Regression কাজ করার জন্য একটা pattern (সম্পর্ক) দরকার

👉 No pattern = no learning

Simple Linear Regression: Simple Linear Regression is used to find the relationship between one input (X) and one output (Y) using a straight line

y = α + βx

y = output (dependent variable)

x = input (independent variable)

α (alpha) = intercept (starting point)

β (beta) = slope (rate of change)

```python
from sklearn.linear_model import LinearRegression

X = [[1], [2], [3]]
y = [2, 4, 6]

model = LinearRegression()
model.fit(X, y)

print(model.predict([[6]]))
```

Getting Started with Linear Regression

#importing libraries

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_squared_error

import math

DataFraming

```python
dataset=pd.read_csv("headbrain.csv")
dataset.head()
```

```python
size = dataset['Head Size(cm^3)']
weight = dataset['Brain Weight(grams)']

print(size)
print(weight)
```

Visualizing Dataset

```python
plt.scatter(size, weight)
plt.show()
```

Converting dataframes into NumPy Arrays

```python
X = np.array(size).reshape(-1,1)
y = np.array(weight).reshape(-1,1)
```

Training the Model

```python
model = LinearRegression()
model.fit(X,y)
```

Mean Sqaured Error & R - Squared Value

1️⃣ Mean Squared Error (MSE)
👉 What is MSE?

👉 English:
MSE measures the average squared error between actual and predicted values

👉 বাংলা:
MSE হলো আসল মান আর predicted মানের পার্থক্যের গড় (square করে)

🔹 Formula
𝑀𝑆𝐸=1/𝑛∑(𝑦−𝑦^)2

🔹 Example

👉 Actual: [2, 4, 6]
👉 Predicted: [2, 3, 7]

👉 Errors:

0, 1, -1

👉 Square:

0, 1, 1

👉 MSE:

(0+1+1)/3 = 0.67

🔹 Key Idea

👉 English:
Lower MSE = better model

👉 বাংলা:
MSE কম হলে → মডেল ভালো

🔹 2️⃣ R-Squared (R²)
👉 What is R²?

👉 English:
R² shows how well the model explains the data

👉 বাংলা:
R² বলে মডেল ডেটা কতটা ভালোভাবে explain করছে

🔹 Range

👉 English:

1 → perfect model
0 → no improvement
<0 → bad model

🔹 Example

👉 R² = 0.9
👉 English: 90% data explained
👉 বাংলা: ৯০% ডেটা explain হয়েছে

```python
regression_model_mse = mean_squared_error(X,y)
print('Mean Squared Error:\t',math.sqrt(regression_model_mse))
print("R squared value\t\t",model.score(X,y))
```

```python
print(model.coef_[0])
print(model.intercept_[0])
```

Visualization

```python
plt.scatter(X,y, color = 'orange')
plt.plot(X,model.predict(X), color = 'black')
plt.title('Linear Regression')
plt.xlabel('Head Size')
plt.ylabel('Brain Weight')
```

Prediction

```python
print('Prediction by the model:\t', model.predict([[2000]]))
```
