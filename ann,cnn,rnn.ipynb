# ANN (Artificial Neural Network)

==================================================

PROBLEM:

Student pass করবে নাকি fail predict করতে হবে।

--------------------------------------------------

INPUT FEATURES

x1 = Study Hour
x2 = Attendance

--------------------------------------------------

ACTUAL INPUT VALUES

x1 = 2
x2 = 50

মানে:

Study Hour = 2 hour
Attendance = 50%

--------------------------------------------------

TARGET OUTPUT

y = 1

মানে:
1 = Pass
0 = Fail

==================================================
STEP 1: INITIALIZE WEIGHTS

Weight মানে input কত important সেটা।

--------------------------------------------------

w1 = Weight for Study Hour
w2 = Weight for Attendance

--------------------------------------------------

INITIAL WEIGHT VALUES

w1 = 0.01
w2 = 0.02

মানে:

Study Hour এর importance = 0.01
Attendance এর importance = 0.02

এখানে attendance বেশি important।

==================================================
STEP 2: INITIALIZE BIAS

Bias হলো extra adjustable value।

--------------------------------------------------

b = 0.1

==================================================
STEP 3: FORWARD PROPAGATION

এখন ANN prediction করবে।

--------------------------------------------------

A) WEIGHTED SUM

Formula:

z = (x1 * w1) + (x2 * w2) + b

--------------------------------------------------

Value বসাই:

z = (2 * 0.01) + (50 * 0.02) + 0.1

--------------------------------------------------

Calculation:

= 0.02 + 1 + 0.1

= 1.12

--------------------------------------------------

FINAL:

z = 1.12

==================================================
STEP 4: ACTIVATION FUNCTION

এখন sigmoid ব্যবহার হবে।

--------------------------------------------------

Formula:

sigmoid(z) = 1 / (1 + e^(-z))

--------------------------------------------------

Value বসাই:

sigmoid(1.12)

--------------------------------------------------

Calculation:

e^(-1.12) ≈ 0.326

1 + 0.326 = 1.326

1 / 1.326 = 0.754

--------------------------------------------------

FINAL OUTPUT:

y_hat = 0.754

==================================================
STEP 5: PREDICTION

Rule:

If y_hat > 0.5
→ PASS

If y_hat < 0.5
→ FAIL

--------------------------------------------------

Here:

0.754 > 0.5

--------------------------------------------------

PREDICTION:

PASS

==================================================
STEP 6: LOSS CALCULATION

Model কত ভুল করেছে সেটা বের হবে।

--------------------------------------------------

ACTUAL OUTPUT:

y = 1

PREDICTED OUTPUT:

y_hat = 0.754

--------------------------------------------------

Formula:

Loss = (y - y_hat)^2

--------------------------------------------------

Value বসাই:

Loss = (1 - 0.754)^2

--------------------------------------------------

Calculation:

= (0.246)^2

= 0.0605

--------------------------------------------------

FINAL LOSS:

Loss = 0.0605

==================================================
STEP 7: ERROR CALCULATION

Formula:

Error = y - y_hat

--------------------------------------------------

Value বসাই:

Error = 1 - 0.754

--------------------------------------------------

Calculation:

= 0.246

--------------------------------------------------

FINAL ERROR:

Error = 0.246

==================================================
STEP 8: LEARNING RATE

Learning Rate decide করে
কত দ্রুত weight update হবে।

--------------------------------------------------

learning_rate = 0.1

==================================================
STEP 9: UPDATE WEIGHT w1

Formula:

w1_new = w1_old + (learning_rate * Error * x1)

--------------------------------------------------

Value বসাই:

w1_new = 0.01 + (0.1 * 0.246 * 2)

--------------------------------------------------

Calculation:

= 0.01 + 0.0492

= 0.0592

--------------------------------------------------

UPDATED w1:

w1 = 0.0592

==================================================
STEP 10: UPDATE WEIGHT w2

Formula:

w2_new = w2_old + (learning_rate * Error * x2)

--------------------------------------------------

Value বসাই:

w2_new = 0.02 + (0.1 * 0.246 * 50)

--------------------------------------------------

Calculation:

= 0.02 + 1.23

= 1.25

--------------------------------------------------

UPDATED w2:

w2 = 1.25

==================================================
STEP 11: UPDATE BIAS

Formula:

b_new = b_old + (learning_rate * Error)

--------------------------------------------------

Value বসাই:

b_new = 0.1 + (0.1 * 0.246)

--------------------------------------------------

Calculation:

= 0.1 + 0.0246

= 0.1246

--------------------------------------------------

UPDATED BIAS:

b = 0.1246

==================================================
STEP 12: AGAIN FORWARD PROPAGATION

নতুন weight দিয়ে আবার prediction হবে।

--------------------------------------------------

Formula:

z = (x1 * w1) + (x2 * w2) + b

--------------------------------------------------

Value বসাই:

z = (2 * 0.0592) + (50 * 1.25) + 0.1246

--------------------------------------------------

Calculation:

= 0.1184 + 62.5 + 0.1246

= 62.743

--------------------------------------------------

NEW z:

62.743

==================================================
STEP 13: AGAIN ACTIVATION

sigmoid(62.743)

--------------------------------------------------

Result:

≈ 1

--------------------------------------------------

NEW PREDICTION:

PASS with almost 100% confidence

==================================================
STEP 14: AGAIN LOSS

Loss = (1 - 1)^2

--------------------------------------------------

= 0

==================================================
FINAL RESULT

Model:

1. Input নিলো
2. Weight multiply করলো
3. Bias add করলো
4. Prediction দিলো
5. Loss বের করলো
6. Error বের করলো
7. Weight update করলো
8. আবার prediction দিলো
9. Loss কমে গেল

==================================================
THIS IS HOW ANN LEARNS

```python
# ==================================================
# ANN Mathematical Calculation
# ==================================================

import numpy as np

# --------------------------
# INPUT FEATURES
# --------------------------

x1 = 2          # Study Hour
x2 = 50         # Attendance

# --------------------------
# TARGET OUTPUT
# --------------------------

y = 1           # Pass

# --------------------------
# INITIAL WEIGHTS
# --------------------------

w1 = 0.01
w2 = 0.02

# --------------------------
# BIAS
# --------------------------

b = 0.1

# --------------------------
# LEARNING RATE
# --------------------------

learning_rate = 0.1

# ==================================================
# STEP 1: WEIGHTED SUM
# z = (x1*w1) + (x2*w2) + b
# ==================================================

z = (x1 * w1) + (x2 * w2) + b

print("Weighted Sum (z):", z)

# ==================================================
# STEP 2: SIGMOID ACTIVATION
# y_hat = 1 / (1 + e^(-z))
# ==================================================

y_hat = 1 / (1 + np.exp(-z))

print("Prediction Probability:", y_hat)

# ==================================================
# STEP 3: FINAL PREDICTION
# ==================================================

if y_hat > 0.5:
    print("Prediction: PASS")
else:
    print("Prediction: FAIL")

# ==================================================
# STEP 4: LOSS CALCULATION
# Loss = (y - y_hat)^2
# ==================================================

loss = (y - y_hat) ** 2

print("Loss:", loss)

# ==================================================
# STEP 5: ERROR CALCULATION
# Error = y - y_hat
# ==================================================

error = y - y_hat

print("Error:", error)

# ==================================================
# STEP 6: UPDATE WEIGHTS
# ==================================================

new_w1 = w1 + (learning_rate * error * x1)

new_w2 = w2 + (learning_rate * error * x2)

# ==================================================
# STEP 7: UPDATE BIAS
# ==================================================

new_b = b + (learning_rate * error)

print("\nUpdated Parameters")
print("Updated w1:", new_w1)
print("Updated w2:", new_w2)
print("Updated b :", new_b)

# ==================================================
# STEP 8: AGAIN FORWARD PROPAGATION
# ==================================================

new_z = (x1 * new_w1) + (x2 * new_w2) + new_b

new_y_hat = 1 / (1 + np.exp(-new_z))

new_loss = (y - new_y_hat) ** 2

print("\nAfter Training")

print("New Weighted Sum:", new_z)
print("New Prediction:", new_y_hat)
print("New Loss:", new_loss)
```

```python
# ==================================================
# ANN MODEL BUILD USING TENSORFLOW / KERAS
# ==================================================

import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# ==================================================
# DATASET
# ==================================================

# Input Features:
# [Study Hour, Attendance]

X = np.array([
    [2, 50],
    [5, 70],
    [8, 90],
    [1, 30],
    [6, 80]
])

# Target Output
# 1 = Pass
# 0 = Fail

y = np.array([0, 1, 1, 0, 1])

# ==================================================
# BUILD ANN MODEL
# ==================================================

model = Sequential()

# Hidden Layer
model.add(Dense(
    units=4,
    activation='relu',
    input_shape=(2,)
))

# Output Layer
model.add(Dense(
    units=1,
    activation='sigmoid'
))

# ==================================================
# COMPILE MODEL
# ==================================================

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# ==================================================
# TRAIN MODEL
# ==================================================

model.fit(
    X,
    y,
    epochs=100
)

# ==================================================
# TEST PREDICTION
# ==================================================

test_data = np.array([[7, 85]])

prediction = model.predict(test_data)

print("\nPrediction Probability:", prediction)

# ==================================================
# FINAL RESULT
# ==================================================

if prediction > 0.5:
    print("Prediction: PASS")
else:
    print("Prediction: FAIL")
```

CNN (Convolutional Neural Network)

CNN mainly image processing এর জন্য ব্যবহার হয়।

    ↓

```python
# ==================================================
# CNN Mathematical Calculation
# ==================================================

import numpy as np

# ==================================================
# INPUT IMAGE
# ==================================================

image = np.array([
    [1, 2, 0],
    [4, 3, 1],
    [1, 2, 2]
])

# ==================================================
# KERNEL / FILTER
# ==================================================

kernel = np.array([
    [1, 0],
    [0, 1]
])

# ==================================================
# CONVOLUTION
# ==================================================

feature_map = np.zeros((2, 2))

for i in range(2):
    for j in range(2):

        region = image[i:i+2, j:j+2]

        feature_map[i, j] = np.sum(region * kernel)

print("Feature Map:")
print(feature_map)

# ==================================================
# RELU ACTIVATION
# ==================================================

relu_output = np.maximum(0, feature_map)

print("\nReLU Output:")
print(relu_output)

# ==================================================
# MAX POOLING
# ==================================================

pooled_output = np.max(relu_output)

print("\nPooled Output:")
print(pooled_output)

# ==================================================
# FLATTEN
# ==================================================

flatten = pooled_output

# ==================================================
# DENSE LAYER
# ==================================================

weight = 0.5
bias = 0.1

z = (flatten * weight) + bias

# ==================================================
# SIGMOID ACTIVATION
# ==================================================

prediction = 1 / (1 + np.exp(-z))

print("\nPrediction Probability:")
print(prediction)

# ==================================================
# FINAL PREDICTION
# ==================================================

if prediction > 0.5:
    print("\nPrediction: CAT")
else:
    print("\nPrediction: DOG")

# ==================================================
# LOSS CALCULATION
# ==================================================

y = 1

loss = (y - prediction) ** 2

print("\nLoss:")
print(loss)
```

```python
# ==================================================
# CNN MODEL BUILD USING TENSORFLOW / KERAS
# ==================================================

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense

# ==================================================
# BUILD CNN MODEL
# ==================================================

model = Sequential()

# ==================================================
# FIRST CONVOLUTION LAYER
# ==================================================

model.add(Conv2D(
    filters=32,
    kernel_size=(3,3),
    activation='relu',
    input_shape=(128,128,3)
))

# ==================================================
# FIRST MAX POOLING
# ==================================================

model.add(MaxPooling2D(
    pool_size=(2,2)
))

# ==================================================
# SECOND CONVOLUTION LAYER
# ==================================================

model.add(Conv2D(
    filters=64,
    kernel_size=(3,3),
    activation='relu'
))

# ==================================================
# SECOND MAX POOLING
# ==================================================

model.add(MaxPooling2D(
    pool_size=(2,2)
))

# ==================================================
# FLATTEN LAYER
# ==================================================

model.add(Flatten())

# ==================================================
# DENSE LAYER
# ==================================================

model.add(Dense(
    units=128,
    activation='relu'
))

# ==================================================
# OUTPUT LAYER
# ==================================================

model.add(Dense(
    units=1,
    activation='sigmoid'
))

# ==================================================
# COMPILE MODEL
# ==================================================

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# ==================================================
# MODEL SUMMARY
# ==================================================

model.summary()
```

RNN Main Formula
h_t = tanh((Wx * x_t) + (Wh * h_(t-1)) + b)

x_t      = Current Input

b        = Bias

```python
# ==================================================
# RNN Mathematical Calculation (Manual)
# ==================================================

import numpy as np

# ==================================================
# WORD ENCODING
# ==================================================

x1 = 1      # "I"
x2 = 2      # "love"

# ==================================================
# INITIAL PARAMETERS
# ==================================================

Wx = 0.5
Wh = 0.8
b = 0.1

# ==================================================
# INITIAL HIDDEN STATE
# ==================================================

h0 = 0

# ==================================================
# FIRST TIME STEP
# ==================================================

h1 = np.tanh((Wx * x1) + (Wh * h0) + b)

print("Hidden State h1:", h1)

# ==================================================
# SECOND TIME STEP
# ==================================================

h2 = np.tanh((Wx * x2) + (Wh * h1) + b)

print("Hidden State h2:", h2)

# ==================================================
# OUTPUT LAYER
# ==================================================

output_weight = 1.2
output_bias = 0.2

z = (h2 * output_weight) + output_bias

print("\nWeighted Sum z:", z)

# ==================================================
# SIGMOID ACTIVATION
# ==================================================

prediction = 1 / (1 + np.exp(-z))

print("Prediction Probability:", prediction)

# ==================================================
# FINAL PREDICTION
# ==================================================

if prediction > 0.5:
    print("\nPrediction: Bangladesh")
else:
    print("\nPrediction: Unknown")

# ==================================================
# LOSS CALCULATION
# ==================================================

y = 1

loss = (y - prediction) ** 2

print("\nLoss:", loss)
```

```python
# ==================================================
# RNN MODEL BUILD USING TENSORFLOW / KERAS
# ==================================================

import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN
from tensorflow.keras.layers import Dense

# ==================================================
# DATASET
# ==================================================

X = np.array([
    [[1], [2]],
    [[2], [3]],
    [[3], [4]],
    [[4], [5]]
])

y = np.array([3, 4, 5, 6])

# ==================================================
# BUILD RNN MODEL
# ==================================================

model = Sequential()

# ==================================================
# RNN LAYER
# ==================================================

model.add(SimpleRNN(
    units=8,
    activation='tanh',
    input_shape=(2,1)
))

# ==================================================
# OUTPUT LAYER
# ==================================================

model.add(Dense(1))

# ==================================================
# COMPILE MODEL
# ==================================================

model.compile(
    optimizer='adam',
    loss='mse'
)

# ==================================================
# TRAIN MODEL
# ==================================================

model.fit(
    X,
    y,
    epochs=100
)

# ==================================================
# TEST PREDICTION
# ==================================================

test = np.array([[[5], [6]]])

prediction = model.predict(test)

print("\nPredicted Next Number:")
print(prediction)
```
